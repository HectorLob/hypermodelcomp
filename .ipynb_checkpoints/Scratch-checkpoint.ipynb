## Demonstrate path independence

```python
import numpy as np
import matplotlib.pyplot as plt
from material_models import Fung, GOH, Fung3D
from NODE_fns import forward_pass
from NODE_fns import NODE as NN
import pickle
import matplotlib
font = {'size'   : 16}
matplotlib.rc('font', **font)
```

```python
with open('savednet/P1C1_GOH.npy', 'rb') as f:
    GOH_wb = pickle.load(f)
```

```python
n = 100
r = np.linspace(0,1,n)
I1_1 = r
I1_2 = r**2
I1_3 = r**(1/2)

Psi_1 = np.zeros(n)
Psi_2 = np.zeros(n)
Psi_3 = np.zeros(n)
```

```python
for i in range(1,n):
    dummyI1 = I1_1[i]
    dummyI2 = 0.
    dummyIv = 0.
    dummyIw = 0.
    dI1 = I1_1[i]-I1_1[i-1]
    I1_params, I2_params, Iv_params, Iw_params, J1_params, J2_params, J3_params, J4_params, J5_params, J6_params, \
        I_weights, theta, Psi1_bias, Psi2_bias = GOH_wb

    a = 1/(1+np.exp(-I_weights))

    J1 = a[0]*dummyI1+(1-a[0])*dummyI2
    J2 = a[1]*dummyI1+(1-a[1])*dummyIv
    J3 = a[2]*dummyI1+(1-a[2])*dummyIw

    Psi1 = NN(dummyI1,  I1_params)
    Psi2 = NN(dummyI2,  I2_params)
    Phi1 = NN(J1,  J1_params)
    Phi2 = NN(J2,  J2_params)
    Phi3 = NN(J3,  J3_params)

    Phi1 = np.max([Phi1, 0])
    Phi2 = np.max([Phi2, 0])
    Phi3 = np.max([Phi3, 0])

    Psi1 = Psi1 + a[0]*Phi1 + a[1]*Phi2 + a[2]*Phi3 + np.exp(Psi1_bias)
    Psi_1[i] = Psi_1[i-1] + Psi1*dI1
        
```

```python
for i in range(1,n):
    dummyI1 = I1_2[i]
    dummyI2 = 0.
    dummyIv = 0.
    dummyIw = 0.
    dI1 = I1_2[i]-I1_2[i-1]
    I1_params, I2_params, Iv_params, Iw_params, J1_params, J2_params, J3_params, J4_params, J5_params, J6_params, \
        I_weights, theta, Psi1_bias, Psi2_bias = GOH_wb

    a = 1/(1+np.exp(-I_weights))

    J1 = a[0]*dummyI1+(1-a[0])*dummyI2
    J2 = a[1]*dummyI1+(1-a[1])*dummyIv
    J3 = a[2]*dummyI1+(1-a[2])*dummyIw

    Psi1 = NN(dummyI1,  I1_params)
    Psi2 = NN(dummyI2,  I2_params)
    Phi1 = NN(J1,  J1_params)
    Phi2 = NN(J2,  J2_params)
    Phi3 = NN(J3,  J3_params)

    Phi1 = np.max([Phi1, 0])
    Phi2 = np.max([Phi2, 0])
    Phi3 = np.max([Phi3, 0])

    Psi1 = Psi1 + a[0]*Phi1 + a[1]*Phi2 + a[2]*Phi3 + np.exp(Psi1_bias)
    Psi_2[i] = Psi_2[i-1] + Psi1*dI1
```

```python
for i in range(1,n):
    dummyI1 = I1_3[i]
    dummyI2 = 0.
    dummyIv = 0.
    dummyIw = 0.
    dI1 = I1_3[i]-I1_3[i-1]
    I1_params, I2_params, Iv_params, Iw_params, J1_params, J2_params, J3_params, J4_params, J5_params, J6_params, \
        I_weights, theta, Psi1_bias, Psi2_bias = GOH_wb

    a = 1/(1+np.exp(-I_weights))

    J1 = a[0]*dummyI1+(1-a[0])*dummyI2
    J2 = a[1]*dummyI1+(1-a[1])*dummyIv
    J3 = a[2]*dummyI1+(1-a[2])*dummyIw

    Psi1 = NN(dummyI1,  I1_params)
    Psi2 = NN(dummyI2,  I2_params)
    Phi1 = NN(J1,  J1_params)
    Phi2 = NN(J2,  J2_params)
    Phi3 = NN(J3,  J3_params)

    Phi1 = np.max([Phi1, 0])
    Phi2 = np.max([Phi2, 0])
    Phi3 = np.max([Phi3, 0])

    Psi1 = Psi1 + a[0]*Phi1 + a[1]*Phi2 + a[2]*Phi3 + np.exp(Psi1_bias)
    Psi_3[i] = Psi_3[i-1] + Psi1*dI1
```

```python
fig, ax = plt.subplots(1,2, figsize=[12,5])
fig.subplots_adjust(wspace=0.4)
ax[0].plot(r, I1_1, label='$I_1 = r$')
ax[0].plot(r, I1_2, label='$I_1 = r^2$')
ax[0].plot(r, I1_3, label='$I_1 = \sqrt{r}$')
ax[0].set(ylabel = '$I_1$', xlabel='r')
ax[0].legend()

ax[1].plot(r, Psi_1[:])
ax[1].plot(r, Psi_2[:])
ax[1].plot(r, Psi_3[:])
ax[1].set(ylabel='$\Psi$', xlabel='r')
pass
```

```python
n = 100
r = np.linspace(0,1,n)
I1_1 = r
I1_2 = r**2
I1_3 = r**(1/2)
I2_1 = r
I2_2 = r**2
I2_3 = r**(1/2)
Iv_1 = r
Iv_2 = r**2
Iv_3 = r**(1/2)
Iw_1 = r
Iw_2 = r**2
Iw_3 = r**(1/2)

Psi_1 = np.zeros(n)
Psi_2 = np.zeros(n)
Psi_3 = np.zeros(n)
```

```python
for i in range(n):
    dummyI1 = I1_1[i]
    dummyI2 = I2_1[i]
    dummyIv = Iv_1[i]
    dummyIw = Iw_1[i]
    dI1 = I1_1[i]-I1_1[i-1]
    dI2 = I2_1[i]-I2_1[i-1]
    dIv = Iv_1[i]-Iv_1[i-1]
    dIw = Iw_1[i]-Iw_1[i-1]
    I1_params, I2_params, Iv_params, Iw_params, J1_params, J2_params, J3_params, J4_params, J5_params, J6_params, \
        I_weights, theta, Psi1_bias, Psi2_bias = GOH_wb

    a = 1/(1+np.exp(-I_weights))

    J1 = a[0]*dummyI1+(1-a[0])*dummyI2
    J2 = a[1]*dummyI1+(1-a[1])*dummyIv
    J3 = a[2]*dummyI1+(1-a[2])*dummyIw
    J4 = a[3]*dummyI2+(1-a[3])*dummyIv
    J5 = a[4]*dummyI2+(1-a[4])*dummyIw
    J6 = a[5]*dummyIv+(1-a[5])*dummyIw

    Psi1 = NN(dummyI1,  I1_params)
    Psi2 = NN(dummyI2,  I2_params)
    Psiv = NN(dummyIv,  Iv_params)
    Psiw = NN(dummyIw,  Iw_params)
    Phi1 = NN(J1,  J1_params)
    Phi2 = NN(J2,  J2_params)
    Phi3 = NN(J3,  J3_params)
    Phi4 = NN(J4,  J4_params)
    Phi5 = NN(J5,  J5_params)
    Phi6 = NN(J6,  J6_params)

    Psiv = np.max([Psiv, 0])
    Psiw = np.max([Psiw, 0])
    Phi1 = np.max([Phi1, 0])
    Phi2 = np.max([Phi2, 0])
    Phi3 = np.max([Phi3, 0])
    Phi4 = np.max([Phi4, 0])
    Phi5 = np.max([Phi5, 0])
    Phi6 = np.max([Phi6, 0])

    Psi1 = Psi1 +     a[0]*Phi1 +     a[1]*Phi2 +     a[2]*Phi3 + np.exp(Psi1_bias)
    Psi2 = Psi2 + (1-a[0])*Phi1 +     a[3]*Phi4 +     a[4]*Phi5 + np.exp(Psi2_bias)
    Psiv = Psiv + (1-a[1])*Phi2 + (1-a[3])*Phi4 +     a[5]*Phi6
    Psiw = Psiw + (1-a[2])*Phi3 + (1-a[4])*Phi5 + (1-a[5])*Phi6
    Psi_1[i] = Psi_1[i-1] + Psi1*dI1 + Psi2*dI2 + Psiv*dIv + Psiw*dIw
        
```
