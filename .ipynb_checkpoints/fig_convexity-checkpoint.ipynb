```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import ticker, cm
from material_models import Fung, GOH, Fung3D
from NODE_fns import forward_pass
from NODE_fns import NODE as NN
import pickle
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
#Load the NODE
with open('savednet/P1C1_Fung.npy', 'rb') as f:
    Fung_wb = pickle.load(f)
with open('savednet/P1C1_GOH.npy', 'rb') as f:
    GOH_wb = pickle.load(f)
```

```python
def matrix_from_stretch(lm1, lm2, lm3):
    zeros = np.zeros(lm1.shape)
    F = np.array([[lm1, zeros, zeros], [zeros, lm2, zeros], [zeros, zeros, lm3]]).transpose((2,0,1))
    C = np.einsum('...ij,...jk->...ik', F, F.transpose(0,2,1))
    return F, C
```

```python
lm = np.linspace(0.8, 1.25, 50)
lm1, lm2 = np.meshgrid(lm, lm)
lm3 = 1/(lm1*lm2)
F, C = matrix_from_stretch(lm1.flatten(), lm2.flatten(), lm3.flatten())
E = 0.5*(C-np.eye(3))

I1 = np.trace(C, axis1 = 1, axis2 = 2)
C2 = np.einsum('...ij,...jk->...ik', C, C)
I2 = 0.5*(I1**2 - np.trace(C2, axis1 = 1, axis2 = 2))
theta = GOH_wb[-3]
v0 = np.array([ np.cos(theta), np.sin(theta), 0])
w0 = np.array([-np.sin(theta), np.cos(theta), 0])
V0 = np.outer(v0, v0)
W0 = np.outer(w0, w0)
Iv_1 = np.einsum('pij,ij->p', C, V0)
Iw_1 = np.einsum('pij,ij->p', C, W0)

theta = Fung_wb[-3]
v0 = np.array([ np.cos(theta), np.sin(theta), 0])
w0 = np.array([-np.sin(theta), np.cos(theta), 0])
V0 = np.outer(v0, v0)
W0 = np.outer(w0, w0)
Iv_2 = np.einsum('pij,ij->p', C, V0)
Iw_2 = np.einsum('pij,ij->p', C, W0)

I1hat = I1-3
I2hat = I2-3
Ivhat_1 = Iv_1-1
Iwhat_1 = Iw_1-1
Ivhat_2 = Iv_2-1
Iwhat_2 = Iw_2-1
#Integrate to get Psi
Psi_NODE_Fung = np.zeros_like(I1)
Psi_NODE_GOH = np.zeros_like(I1)
```

```python
for i in range(Psi_NODE_Fung.shape[0]):
    npoint = 20
    dI1 = I1hat[i]/npoint
    dI2 = I2hat[i]/npoint
    dIv_1 = Ivhat_1[i]/npoint
    dIw_1 = Iwhat_1[i]/npoint
    dIv_2 = Ivhat_2[i]/npoint
    dIw_2 = Iwhat_2[i]/npoint
    
    dummyI1 = 0.0
    dummyI2 = 0.0
    dummyIv_1 = 0.0
    dummyIw_1 = 0.0
    dummyIv_2 = 0.0
    dummyIw_2 = 0.0
    for j in range(npoint):
        I1_params, I2_params, Iv_params, Iw_params, J1_params, J2_params, J3_params, J4_params, J5_params, J6_params, \
            I_weights, theta, Psi1_bias, Psi2_bias = GOH_wb

        a = 1/(1+np.exp(-I_weights))
        
        J1 = a[0]*dummyI1+(1-a[0])*dummyI2
        J2 = a[1]*dummyI1+(1-a[1])*dummyIv_1
        J3 = a[2]*dummyI1+(1-a[2])*dummyIw_1
        J4 = a[3]*dummyI2+(1-a[3])*dummyIv_1
        J5 = a[4]*dummyI2+(1-a[4])*dummyIw_1
        J6 = a[5]*dummyIv_1+(1-a[5])*dummyIw_1

        Psi1 = NN(dummyI1,  I1_params)
        Psi2 = NN(dummyI2,  I2_params)
        Psiv = NN(dummyIv_1,  Iv_params)
        Psiw = NN(dummyIw_1,  Iw_params)
        Phi1 = NN(J1,  J1_params)
        Phi2 = NN(J2,  J2_params)
        Phi3 = NN(J3,  J3_params)
        Phi4 = NN(J4,  J4_params)
        Phi5 = NN(J5,  J5_params)
        Phi6 = NN(J6,  J6_params)

        Psiv = np.max([Psiv, 0])
        Psiw = np.max([Psiw, 0])
        Phi1 = np.max([Phi1, 0])
        Phi2 = np.max([Phi2, 0])
        Phi3 = np.max([Phi3, 0])
        Phi4 = np.max([Phi4, 0])
        Phi5 = np.max([Phi5, 0])
        Phi6 = np.max([Phi6, 0])

        Psi1 = Psi1 +     a[0]*Phi1 +     a[1]*Phi2 +     a[2]*Phi3 + np.exp(Psi1_bias)
        Psi2 = Psi2 + (1-a[0])*Phi1 +     a[3]*Phi4 +     a[4]*Phi5 + np.exp(Psi2_bias)
        Psiv = Psiv + (1-a[1])*Phi2 + (1-a[3])*Phi4 +     a[5]*Phi6
        Psiw = Psiw + (1-a[2])*Phi3 + (1-a[4])*Phi5 + (1-a[5])*Phi6
        Psi_NODE_GOH[i]+= Psi1*dI1 + Psi2*dI2 + Psiv*dIv_1 + Psiw*dIw_1
        
        
        I1_params, I2_params, Iv_params, Iw_params, J1_params, J2_params, J3_params, J4_params, J5_params, J6_params, \
            I_weights, theta, Psi1_bias, Psi2_bias = Fung_wb

        a = 1/(1+np.exp(-I_weights))
        
        J1 = a[0]*dummyI1+(1-a[0])*dummyI2
        J2 = a[1]*dummyI1+(1-a[1])*dummyIv_2
        J3 = a[2]*dummyI1+(1-a[2])*dummyIw_2
        J4 = a[3]*dummyI2+(1-a[3])*dummyIv_2
        J5 = a[4]*dummyI2+(1-a[4])*dummyIw_2
        J6 = a[5]*dummyIv_2+(1-a[5])*dummyIw_2

        Psi1 = NN(dummyI1,  I1_params)
        Psi2 = NN(dummyI2,  I2_params)
        Psiv = NN(dummyIv_2,  Iv_params)
        Psiw = NN(dummyIw_2,  Iw_params)
        Phi1 = NN(J1,  J1_params)
        Phi2 = NN(J2,  J2_params)
        Phi3 = NN(J3,  J3_params)
        Phi4 = NN(J4,  J4_params)
        Phi5 = NN(J5,  J5_params)
        Phi6 = NN(J6,  J6_params)

        Psiv = np.max([Psiv, 0])
        Psiw = np.max([Psiw, 0])
        Phi1 = np.max([Phi1, 0])
        Phi2 = np.max([Phi2, 0])
        Phi3 = np.max([Phi3, 0])
        Phi4 = np.max([Phi4, 0])
        Phi5 = np.max([Phi5, 0])
        Phi6 = np.max([Phi6, 0])

        Psi1 = Psi1 +     a[0]*Phi1 +     a[1]*Phi2 +     a[2]*Phi3 + np.exp(Psi1_bias)
        Psi2 = Psi2 + (1-a[0])*Phi1 +     a[3]*Phi4 +     a[4]*Phi5 + np.exp(Psi2_bias)
        Psiv = Psiv + (1-a[1])*Phi2 + (1-a[3])*Phi4 +     a[5]*Phi6
        Psiw = Psiw + (1-a[2])*Phi3 + (1-a[4])*Phi5 + (1-a[5])*Phi6
        Psi_NODE_Fung[i]+= Psi1*dI1 + Psi2*dI2 + Psiv*dIv_2 + Psiw*dIw_2
        
        dummyI1+= dI1
        dummyI2+= dI2
        dummyIv_1+= dIv_1
        dummyIw_1+= dIw_1
        dummyIv_2+= dIv_2
        dummyIw_2+= dIw_2
```

```python
F = F.reshape([50,50,3,3])
C = C.reshape([50,50,3,3])
E = E.reshape([50,50,3,3])
Psi_NODE_Fung = Psi_NODE_Fung.reshape([50,50])
Psi_NODE_GOH = Psi_NODE_GOH.reshape([50,50])
I1 = I1.reshape([50,50])
I2 = I2.reshape([50,50])
Iv = Iv.reshape([50,50])
Iw = Iw.reshape([50,50])
```
