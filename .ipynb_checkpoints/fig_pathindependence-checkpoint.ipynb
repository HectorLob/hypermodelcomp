## Demonstrate path independence

```python
import numpy as np
import matplotlib.pyplot as plt
from material_models import Fung, GOH, Fung3D
from NODE_fns import forward_pass
from NODE_fns import NODE as NN
import pickle
import matplotlib
font = {'size'   : 16}
matplotlib.rc('font', **font)
```

```python
with open('savednet/P1C1_GOH.npy', 'rb') as f:
    GOH_wb = pickle.load(f)
```

```python
n = 1000
r = np.linspace(0,1,n)
I1_1 = r
I1_2 = r**2
I1_3 = np.sin(np.pi*r/2)

Psi_1 = np.zeros(n)
Psi_2 = np.zeros(n)
Psi_3 = np.zeros(n)
```

```python
for i in range(1,n):
    dummyI1 = I1_1[i]
    dummyI2 = 0.
    dummyIv = 0.
    dummyIw = 0.
    dI1 = I1_1[i]-I1_1[i-1]
    I1_params, I2_params, Iv_params, Iw_params, J1_params, J2_params, J3_params, J4_params, J5_params, J6_params, \
        I_weights, theta, Psi1_bias, Psi2_bias = GOH_wb

    a = 1/(1+np.exp(-I_weights))

    J1 = a[0]*dummyI1+(1-a[0])*dummyI2
    J2 = a[1]*dummyI1+(1-a[1])*dummyIv
    J3 = a[2]*dummyI1+(1-a[2])*dummyIw

    Psi1 = NN(dummyI1,  I1_params)
    Psi2 = NN(dummyI2,  I2_params)
    Phi1 = NN(J1,  J1_params)
    Phi2 = NN(J2,  J2_params)
    Phi3 = NN(J3,  J3_params)

    Phi1 = np.max([Phi1, 0])
    Phi2 = np.max([Phi2, 0])
    Phi3 = np.max([Phi3, 0])

    Psi1 = Psi1 + a[0]*Phi1 + a[1]*Phi2 + a[2]*Phi3 + np.exp(Psi1_bias)
    Psi_1[i] = Psi_1[i-1] + Psi1*dI1
        
```

```python
for i in range(1,n):
    dummyI1 = I1_2[i]
    dummyI2 = 0.
    dummyIv = 0.
    dummyIw = 0.
    dI1 = I1_2[i]-I1_2[i-1]
    I1_params, I2_params, Iv_params, Iw_params, J1_params, J2_params, J3_params, J4_params, J5_params, J6_params, \
        I_weights, theta, Psi1_bias, Psi2_bias = GOH_wb

    a = 1/(1+np.exp(-I_weights))

    J1 = a[0]*dummyI1+(1-a[0])*dummyI2
    J2 = a[1]*dummyI1+(1-a[1])*dummyIv
    J3 = a[2]*dummyI1+(1-a[2])*dummyIw

    Psi1 = NN(dummyI1,  I1_params)
    Psi2 = NN(dummyI2,  I2_params)
    Phi1 = NN(J1,  J1_params)
    Phi2 = NN(J2,  J2_params)
    Phi3 = NN(J3,  J3_params)

    Phi1 = np.max([Phi1, 0])
    Phi2 = np.max([Phi2, 0])
    Phi3 = np.max([Phi3, 0])

    Psi1 = Psi1 + a[0]*Phi1 + a[1]*Phi2 + a[2]*Phi3 + np.exp(Psi1_bias)
    Psi_2[i] = Psi_2[i-1] + Psi1*dI1
```

```python
for i in range(1,n):
    dummyI1 = I1_3[i]
    dummyI2 = 0.
    dummyIv = 0.
    dummyIw = 0.
    dI1 = I1_3[i]-I1_3[i-1]
    I1_params, I2_params, Iv_params, Iw_params, J1_params, J2_params, J3_params, J4_params, J5_params, J6_params, \
        I_weights, theta, Psi1_bias, Psi2_bias = GOH_wb

    a = 1/(1+np.exp(-I_weights))

    J1 = a[0]*dummyI1+(1-a[0])*dummyI2
    J2 = a[1]*dummyI1+(1-a[1])*dummyIv
    J3 = a[2]*dummyI1+(1-a[2])*dummyIw

    Psi1 = NN(dummyI1,  I1_params)
    Psi2 = NN(dummyI2,  I2_params)
    Phi1 = NN(J1,  J1_params)
    Phi2 = NN(J2,  J2_params)
    Phi3 = NN(J3,  J3_params)

    Phi1 = np.max([Phi1, 0])
    Phi2 = np.max([Phi2, 0])
    Phi3 = np.max([Phi3, 0])

    Psi1 = Psi1 + a[0]*Phi1 + a[1]*Phi2 + a[2]*Phi3 + np.exp(Psi1_bias)
    Psi_3[i] = Psi_3[i-1] + Psi1*dI1
```

```python
fig, ax = plt.subplots(1,2, figsize=[12,5])
fig.subplots_adjust(wspace=0.4)
ax[0].plot(r, I1_1, label='$J_1 = r$')
ax[0].plot(r, I1_2, label='$J_1 = r^2$')
ax[0].plot(r, I1_3, label='$J_1 = sin(\pi r/2)$')
ax[0].set(ylabel = '$J_1$', xlabel='r')
ax[0].legend()

ax[1].plot(r, Psi_1)
ax[1].plot(r, Psi_2)
ax[1].plot(r, Psi_3)
ax[1].set(ylabel='$\Psi$', xlabel='r')
pass
fig.savefig('Figures/fig_pathindependence.png', dpi=400, bbox_inches='tight')
```
