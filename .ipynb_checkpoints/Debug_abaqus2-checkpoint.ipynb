```python
import numpy as np
from material_models import GOH, MR, HGO, Fung, neoHook, lm2F
import pickle
import matplotlib.pyplot as plt
from NODE_fns import NODE_sigma_vmap, NODE
```

```python
GOH_params = [1.02356332e-02, 5.13664702e-01, 5.91491834e+01, 2.74447648e-01, 0]#1.57079633e+00]
GOH_MM  = GOH(GOH_params)
```

```python
with open('savednet/' + 'P1C1_GOH_I1Iv.npy', 'rb') as f:
    params = pickle.load(f)
I1_params, I2_params, Iv_params, Iw_params, J1_params, J2_params, J3_params, J4_params, J5_params, J6_params, I_weights, theta, \
    Psi1_bias, Psi2_bias = params
a = 1/(1+np.exp(-I_weights))
```

```python
I1 = np.zeros([1,1])
I2 = np.zeros([1,1])
Iv = np.zeros([1,1])
Iw = np.zeros([1,1])
I1[0,0] = 3.107
I2[0,0] = 3.094
Iv[0,0] = 1.440
Iw[0,0] = 0.833
print('Invariants: (I1, I2, Iv, Iw)')
print(I1[0,0], I2[0,0], Iv[0,0], Iw[0,0])
print('----------------------------')

print('First derivatives with NODE')
I1 = I1-3
I2 = I2-3
Iv = Iv-1
Iw = Iw-1
J1 = a[0]*I1+(1-a[0])*I2
J2 = a[1]*I1+(1-a[1])*Iv
J3 = a[2]*I1+(1-a[2])*Iw
J4 = a[3]*I2+(1-a[3])*Iv
J5 = a[4]*I2+(1-a[4])*Iw
J6 = a[5]*Iv+(1-a[5])*Iw


#Temporary, remove if unsure:
I2, Iw, J1, J3, J4, J5, J6 = 0.,0.,0.,0.,0.,0.,0.
    

Psi1 = NODE(I1,  I1_params)
Psi2 = NODE(I2,  I2_params)
Psiv = NODE(Iv,  Iv_params)
Psiw = NODE(Iw,  Iw_params)
Phi1 = NODE(J1,  J1_params)
Phi2 = NODE(J2,  J2_params)
Phi3 = NODE(J3,  J3_params)
Phi4 = NODE(J4,  J4_params)
Phi5 = NODE(J5,  J5_params)
Phi6 = NODE(J6,  J6_params)

Psiv = np.max([Psiv, 0])
Psiw = np.max([Psiw, 0])
Phi1 = np.max([Phi1, 0])
Phi2 = np.max([Phi2, 0])
Phi3 = np.max([Phi3, 0])
Phi4 = np.max([Phi4, 0])
Phi5 = np.max([Phi5, 0])
Phi6 = np.max([Phi6, 0])

Psi1 = Psi1 +     a[0]*Phi1 +     a[1]*Phi2 +     a[2]*Phi3 + np.exp(Psi1_bias)
Psi2 = Psi2 + (1-a[0])*Phi1 +     a[3]*Phi4 +     a[4]*Phi5 + np.exp(Psi2_bias)
Psiv = Psiv + (1-a[1])*Phi2 + (1-a[3])*Phi4 +     a[5]*Phi6
Psiw = Psiw + (1-a[2])*Phi3 + (1-a[4])*Phi5 + (1-a[5])*Phi6


#Temporary, remove if unsure:
Psi2 = 0.
    
    
print(Psi1[0,0], Psi2, Psiv[0,0], Psiw)#[0,0])
print('----------------------------')
I1 = I1+3
Iv = Iv+1
print('First derivatives with GOH')
Psi1, Psiv = GOH_MM.partials_from_inv(I1,Iv)
print(Psi1[0,0], Psiv[0,0])
```
